```python
sex= pd.get_dummies(Titanic_Data['Sex'], drop_first=True)
sex.head(5)
```

```python
embark=pd.get_dummies(Titanic_Data['Embarked'], drop_first=True)
embark.head(5)
```

```python
class_p=pd.get_dummies(Titanic_Data['Pclass'], drop_first=True)
class_p.head(5)
```

```python
Titanic_Data=pd.concat([Titanic_Data, sex, embark, class_p], axis=1)
Titanic_Data.head(5)
```

```python
Titanic_Data.drop(['PassengerId', 'Sex', 'Name', 'Ticket'], axis=1, inplace=True)
```
